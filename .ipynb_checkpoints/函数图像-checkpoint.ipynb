```python
from draw_fx import draw_fx
import numpy as np
```

二值随机变量香农熵

```python
def H(x):
    return (x-1)*np.log(1-x) - x*np.log(x)
```

```python
draw_fx(H, max_x=0.999, min_x=0.001, title="香农熵")
```

softplus 函数（ReLU函数的平滑版）

```python
def softplus(x):
    return np.log(1 + np.exp(x))
```

```python
draw_fx(softplus, max_x=10, min_x=-10, title="softplus")
```
